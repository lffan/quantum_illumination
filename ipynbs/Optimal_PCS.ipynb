```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
```

```python
from matplotlib import rc
rc('xtick', labelsize=14) 
rc('ytick', labelsize=14) 
np.set_printoptions(threshold='nan', precision=6, suppress=True)
```

```python
df1 = pd.read_csv('../output/expr_2_pcs_nth_0.1_grid101.csv', comment='#')
df2 = pd.read_csv('../output/expr_2_pcs_nth_10_grid10.csv', comment='#')
```

```python
sns.heatmap(df1['Helstrom'].values.reshape(101, 101), 
            cmap="RdBu_r", xticklabels=False, yticklabels=False)
# plt.xlabel("ra")
# plt.ylabel("rb")
plt.show()
```

```python
sns.heatmap(df2['Helstrom'].values.reshape(10, 10), 
            cmap="RdBu_r", xticklabels=False, yticklabels=False)
# plt.xlabel("ra")
# plt.ylabel("rb")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
xs = df1['ra'].values
ys = df1['rb'].values
zs = df1['Helstrom'].values
ax.plot_trisurf(xs, ys, zs, cmap=plt.cm.Spectral)
ax.set_xlabel('$r_a$')
ax.set_ylabel('$r_b$')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
xs = df2['ra'].values
ys = df2['rb'].values
zs = df2['Helstrom'].values
ax.plot_trisurf(xs, ys, zs, cmap=plt.cm.Spectral)
ax.set_xlabel('$r_a$')
ax.set_ylabel('$r_b$')
plt.show()
```

```python
df3 = pd.read_csv('../output/expr_3_lmd_p101.csv', comment='#')
```

```python
for name in ('TMSS', 'PS', 'PA', 'PAS', 'PSA', 'PCS'):
    temp_df = df3[df3['State'] == name]
    plt.plot(temp_df['Aver_N'], temp_df['Helstrom'], 
             lw=2, label=name)
plt.xlim(0, 3)
plt.ylim(0.42, 0.5)
plt.legend(loc=3, fontsize=14)
plt.show()
```

```python
for name in ('TMSS', 'PS', 'PA', 'PAS', 'PSA', 'PCS'):
    temp_df = df3[df3['State'] == name]
    plt.plot(temp_df['Aver_N'], temp_df['Chernoff'], 
             lw=2, label=name)
plt.xlim(0, 3)
plt.ylim(0.49, 0.5)
plt.legend(fontsize=14)
plt.show()
```

```python
for name in ('TMSS', 'PS', 'PA', 'PAS', 'PSA', 'PCS'):
    temp_df = df3[df3['State'] == name]
    plt.plot(temp_df['Aver_N'], temp_df['VN_Entropy'], 
             lw=2, label=name)
plt.xlim(0, 3)
# plt.ylim(0.49, 0.5)
plt.legend(fontsize=14)
plt.show()
```
