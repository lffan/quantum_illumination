```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# configuration
from IPython.display import set_matplotlib_formats
plt.style.use(['seaborn-whitegrid'])
set_matplotlib_formats('pdf', 'png')
np.set_printoptions(threshold='nan', precision=6, suppress=True)
%matplotlib inline
```

```python
df1 = pd.read_csv('../output/data/basic_attributes.csv', comment='#')
names = ('TMSS', 'PS', 'PA', 'PAS', 'PSA', 'PCS')
```

```python
df1.head()
```

```python
for name in ('PS', 'PSA'):
    temp_df = df1[df1['State'] == name]
    plt.plot(temp_df['Aver_N'], temp_df['VN_Entropy'], 
             lw=2, label=name)
for name in ('PA', 'PAS', 'PCS'):
    temp_df = df1[df1['State'] == name]
    plt.plot(temp_df['Aver_N'] - 2, temp_df['VN_Entropy'], 
             lw=2, label=name)
# plt.xlim(0, 2)
# plt.ylim(0.43, 0.5)
plt.legend(loc=0, fontsize=14)
plt.xlabel("Average Photon Number")
plt.ylabel("Entropy")
plt.title("Von Neumann Entropy")
plt.show()
```

```python
for name in names:
    temp_df = df1[df1['State'] == name]
    plt.plot(temp_df['Aver_N'], temp_df['VN_Entropy'], 
             lw=2, label=name)
# plt.xlim(0, 2)
# plt.ylim(0.43, 0.5)
plt.legend(loc=0, fontsize=14)
plt.xlabel("Average Photon Number")
plt.ylabel("Entropy")
plt.title("Von Neumann Entropy")
plt.show()
```

```python
for name in ('TMSS', 'PS', 'PA', 'PAS'):
    temp_df = df1[df1['State'] == name]
    plt.plot(temp_df['Aver_N'], 
             (temp_df['Exact_N'] - temp_df['Aver_N']), 
             lw=2, label=name)
# plt.xlim(0, 2)
# plt.ylim(0.43, 0.5)
plt.legend(loc=0, fontsize=14)
plt.xlabel("Average Photon Number")
plt.ylabel("Exact Photon Number")
plt.title("Exact Photon Number")
plt.show()
```

```python
for name in names:
    temp_df = df1[df1['State'] == name]
    plt.plot(temp_df['lambda'], temp_df['VN_Entropy'], 
             lw=2, label=name)
# plt.xlim(0, 2)
# plt.ylim(0.43, 0.5)
plt.legend(loc=0, fontsize=14)
plt.xlabel("$\lambda$")
plt.ylabel("Entropy")
plt.title("Von Neumann Entropy")
plt.show()
```

```python
for name in names:
    temp_df = df1[df1['State'] == name]
    plt.plot(temp_df['lambda'], temp_df['Aver_N'], 
             lw=2, label=name)
# plt.xlim(0, 2)
# plt.ylim(0.43, 0.5)
plt.legend(loc=0, fontsize=14)
plt.xlabel("$\lambda$")
plt.ylabel("Photon Number")
plt.title("Photon Number")
plt.show()
```
