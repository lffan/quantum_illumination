```python
temp = df[df['lambda'] == df['lambda'].unique()[5]]
vn = temp['VN_Entropy'].values.reshape(51, 51)
plt.imshow(vn, cmap='RdBu_r', interpolation='BiCubic', 
           origin='lower', extent=[0, 1, 0, 1.0])
plt.colorbar()
plt.show()
```

```python
print(temp['lambda'].unique())
vn = temp['VN_Entropy'].values.reshape(size, size)
ras = temp['ra'].values.reshape(size, size)
rbs = temp['rb'].values.reshape(size, size)
ss = temp['sqz'].values.reshape(size, size)
nb = temp['B_aver_N'].values.reshape(size, size)

fig = plt.contour(rbs, ras, vn, cmap='gnuplot', 
                  levels=np.arange(0.55, 0.9, 0.02))
plt.clabel(fig, inline=1, fontsize=12)
plt.axes().set_aspect('equal')
plt.title('PCS - entropy of entanglement')
plt.xlabel('$r$')
plt.ylabel('$s$')
plt.show()
```
