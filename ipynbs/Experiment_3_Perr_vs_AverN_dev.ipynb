```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# configuration
from IPython.display import set_matplotlib_formats
plt.style.use(['seaborn-whitegrid'])
set_matplotlib_formats('pdf', 'png')
np.set_printoptions(threshold='nan', precision=6, suppress=True)
%matplotlib inline
```

```python
df1 = pd.read_csv('../output/data/expr_3_nbar_nth_0.1_g_11_01-31.csv', comment='#')
names = ('TMSS', 'PS', 'PA', 'PAS', 'PSA', 'PCS')
# names = ('TMSS', 'PS', 'PSA')
```

```python
df1.head()
```

```python
for name in names:
    temp_df = df1[df1['State'] == name]
    plt.plot(temp_df['lambda'], temp_df['Aver_N'], label=name)
# plt.xlim(0, 2)
# plt.ylim(0.43, 0.5)
plt.legend(loc=(0.02, 0.1), fontsize=12)
plt.xlabel("$\lambda$")
plt.ylabel("Photon Number")
plt.title("Photon Number")
plt.show()
```

```python
for name in names:
    temp_df = df1[df1['State'] == name]
    avern = temp_df['Aver_N']
    plt.plot(avern, temp_df['Helstrom_Bound'], lw=2, label=name)
# plt.xlim(-0.05, 3)
plt.ylim(0.4, 0.502)
plt.legend(loc=0, fontsize=12)
plt.xlabel("Average Photon Number", fontsize=12)
plt.ylabel("Helstrom Bound", fontsize=12)
plt.title("Helstrom Bound", fontsize=12)
plt.savefig('../output/figures/g_qhb_vs_nbar_nth_1.pdf', bbox_inches='tight')
plt.show()
```

```python
for name in names:
    temp_df = df1[df1['State'] == name]
    if name in ('TMSS', 'PA', 'PAS', 'PCS'):
        avern = temp_df['Aver_N'] / 2
    else:
        avern = temp_df['Aver_N']
    plt.plot(avern, temp_df['Helstrom_Bound'], lw=2, label=name)
plt.xlim(-0.05, 2)
plt.ylim(0.47, 0.502)
plt.legend(loc=0, fontsize=12)
plt.xlabel("Average Photon Number", fontsize=12)
plt.ylabel("Helstrom Bound", fontsize=12)
plt.title("Helstrom Bound", fontsize=12)
plt.savefig('../output/figures/g_qhb_vs_nbar_nth_1.pdf', bbox_inches='tight')
plt.show()
```

```python
for name in ('TMSS', 'PS', 'PSA'):
    temp_df = df1[df1['State'] == name]
    avern = temp_df['Aver_N']
    plt.plot(avern, temp_df['Helstrom_Bound'], lw=2, label=name)
plt.xlim(-0.02, 0.4)
plt.ylim(0.483, 0.501)
plt.legend(loc=3, fontsize=12)
plt.xlabel("Average Photon Number", fontsize=12)
plt.ylabel("Helstrom Bound", fontsize=12)
plt.title("Helstrom Bound under Small Squeezing Parameters", fontsize=12)
plt.legend(loc=0, fontsize=12)
plt.savefig('../output/figures/g_qhb_vs_nbar_nth_1_small.pdf', bbox_inches='tight')
plt.show()
```

```python
for name in ('TMSS', 'PS', 'PSA'):
    temp_df = df1[df1['State'] == name]
    if name in ('TMSS', 'PA', 'PAS', 'PCS'):
        avern = temp_df['Aver_N'] / 2
    else:
        avern = temp_df['Aver_N']
    plt.plot(avern, temp_df['Helstrom_Bound'], lw=2, label=name)
plt.xlim(-0.02, 0.4)
plt.ylim(0.483, 0.501)
plt.legend(loc=3, fontsize=12)
plt.xlabel("Average Photon Number", fontsize=12)
plt.ylabel("Helstrom Bound", fontsize=12)
plt.title("Helstrom Bound under Small Squeezing Parameters", fontsize=12)
plt.legend(loc=0, fontsize=12)
plt.savefig('../output/figures/g_qhb_vs_nbar_nth_1_small.pdf', bbox_inches='tight')
plt.show()
```

```python
for name in names:
    temp_df = df1[df1['State'] == name]
    if name in ('TMSS', 'PA', 'PAS', 'PCS'):
        avern = temp_df['Aver_N'] / 2
    else:
        avern = temp_df['Aver_N']
    plt.plot(avern, temp_df['Chernoff_Bound'], lw=2, label=name)
# plt.xlim(0, 2)
# plt.ylim(0.494, 0.5)
plt.legend(fontsize=14)
plt.xlabel("Average Photon Number", fontsize=14)
plt.ylabel("Chernoff Bound", fontsize=14)
plt.title("Chernoff Bound", fontsize=14)
```

```python
for name in names:
    temp_df = df1[df1['State'] == name]
    avern = temp_df['Aver_N']
    plt.plot(avern, temp_df['VN_Entropy'], lw=2, label=name)
plt.xlim(-0.1, 5)
plt.ylim(0.0, 1.2)
plt.legend(fontsize=12)
plt.xlabel("Average Photon Number", fontsize=12)
plt.ylabel("Entropy", fontsize=12)
plt.title("Entropy", fontsize=12)
plt.savefig('../output/figures/g_entropy_vs_nbar_nth_1.pdf', bbox_inches='tight')
```
