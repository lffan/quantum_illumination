```python
plt.figure(figsize=(6,4))
plt.scatter(dfa['B_aver_N'][::18], dfa['VN_Entropy'][::18], 
            marker='o', s=1.0, alpha=0.5, color='orange', label='PCS')

for name in names:
    temp_df = dfb[dfb['State'] == name]
    if name == 'PCS':
        plt.plot(temp_df['B_aver_N'], temp_df['VN_Entropy'], 
                 dashes=dashes[name], label='PCS-etgl', lw=2)
    elif name == 'TMSS':
        plt.plot(temp_df['B_aver_N'], temp_df['VN_Entropy'], label=name)
    else:
        plt.plot(temp_df['B_aver_N'], temp_df['VN_Entropy'], 
                 dashes=dashes[name], label=name)
plt.xlim(-0.05, 2)
plt.ylim(-0.08, 2.0)
plt.legend(fontsize=13, ncol=2)

# temp = dfa[(dfa['ra'] == 0) & (dfa['rb'] == 0)]
# plt.plot(temp['B_aver_N'], temp['VN_Entropy'])

# # plt.legend()
# plt.xlim(0, 2)
# plt.ylim(0, 2)
plt.xlabel("$\langle n_B \\rangle$")
plt.ylabel("$E$")

plt.savefig('../output/figures/g_entropy_vs_nbarB.pdf', bbox_inches='tight')
plt.show()
```
