```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Read datat

```python
df1 = pd.read_csv('../output/data/basic_attributes.csv', comment='#')
df2 = pd.read_csv('../output/data/basic_attributes_2.csv', comment='#')
names = ('TMSS', 'PS', 'PA', 'PSA', 'PAS', 'PCS')
```

```python
dashes = {'TMSS': [1,1], 'PA':[6,3], 'PS':[3,1], 'PAS':[2,2,1,2], 'PSA':[6,4,2,4], 'PCS':[5,3,1,3]}
```

### 1. Von Neumann Entropy vs. Lambda

```python
for name in names:
    temp_df = df1[df1['State'] == name]
    plt.plot(temp_df['lambda'], temp_df['VN_Entropy'], 
             dashes=dashes[name], label=name)
# plt.xlim(0, 2)
# plt.ylim(0.43, 0.5)
plt.legend(loc=0, fontsize=12)
plt.xlabel("$\lambda$")
plt.ylabel("Entropy")
plt.title("Von Neumann Entropy")
plt.show()
```

```python
for name in names:
    temp_df = df2[df2['State'] == name]
    plt.plot(temp_df['lambda'], temp_df['VN_Entropy'], 
             dashes=dashes[name], label=name)
# plt.xlim(0, 2)
# plt.ylim(0.43, 0.5)
plt.legend(loc=0, fontsize=12)
plt.xlabel("$\lambda$")
plt.ylabel("Entropy")
plt.title("Von Neumann Entropy")
plt.show()
```

### 2. Average Photon Number vs. Lambda

```python
for name in names:
    temp_df = df1[df1['State'] == name]
    plt.plot(temp_df['lambda'], temp_df['Aver_N'], 
             dashes=dashes[name], label=name)
# plt.xlim(0, 2)
# plt.ylim(0.43, 0.5)
plt.legend(loc=0, fontsize=12)
plt.xlabel("$\lambda$")
plt.ylabel("Photon Number")
plt.title("Photon Number")
plt.show()
```

```python
for name in names:
    temp_df = df2[df2['State'] == name]
    plt.plot(temp_df['lambda'], temp_df['Aver_N'], 
             dashes=dashes[name], label=name)
# plt.xlim(0, 2)
# plt.ylim(0.43, 0.5)
plt.legend(loc=(0.02, 0.1), fontsize=12)
plt.xlabel("$\lambda$")
plt.ylabel("Photon Number")
plt.title("Photon Number")
plt.show()
```

### Entropy vs. Average Photon Number

```python
for name in ('TMSS', 'PS', 'PSA'):
    temp_df = df1[df1['State'] == name]
    plt.plot(temp_df['Aver_N'], temp_df['VN_Entropy'], 
             dashes=dashes[name], label=name)
for name in ('PA', 'PAS', 'PCS'):
    temp_df = df1[df1['State'] == name]
    plt.plot((temp_df['Aver_N'] - 0), temp_df['VN_Entropy'], 
             dashes=dashes[name], label=name)
# plt.xlim(0, 2)
# plt.ylim(0.43, 0.5)
plt.legend(loc=0, fontsize=12)
plt.xlabel("Average Photon Number")
plt.ylabel("Entropy")
plt.title("Von Neumann Entropy")
plt.show()
```

```python
for name in ('TMSS', 'PS', 'PSA'):
    temp_df = df2[df2['State'] == name]
    plt.plot(temp_df['Aver_N'], temp_df['VN_Entropy'], 
             dashes=dashes[name], label=name)
for name in ('PA', 'PAS', 'PCS'):
    temp_df = df2[df2['State'] == name]
    plt.plot((temp_df['Aver_N'] - 0), temp_df['VN_Entropy'], 
             dashes=dashes[name], label=name)
# plt.xlim(0, 2)
# plt.ylim(0.43, 0.5)
plt.legend(loc=0, fontsize=12)
plt.xlabel("Average Photon Number")
plt.ylabel("Entropy")
plt.title("Von Neumann Entropy")
plt.show()
```

### Deviation between numberical and analytical average photon numbers

```python
for name in ('TMSS', 'PS', 'PA', 'PAS', 'PCS'):
    temp_df = df2[df2['State'] == name]
    plt.plot(temp_df['Aver_N'], 
             (temp_df['Exact_N'] - temp_df['Aver_N']), 
             dashes=dashes[name], label=name)
# plt.xlim(0, 2)
# plt.ylim(0.43, 0.5)
plt.legend(loc=0, fontsize=14)
plt.xlabel("Average Photon Number")
plt.ylabel("Exact Photon Number")
plt.title("Exact Photon Number")
plt.show()
```

```python
for name in ('PSA',):
    temp_df = df2[df2['State'] == name]
    plt.plot(temp_df['Aver_N'], 
             (temp_df['Exact_N'] - temp_df['Aver_N']), 
             dashes=dashes[name], label=name)
# plt.xlim(0, 2)
# plt.ylim(0.43, 0.5)
plt.legend(loc=0, fontsize=14)
plt.xlabel("Average Photon Number")
plt.ylabel("Exact Photon Number")
plt.title("Exact Photon Number")
plt.show()
```
